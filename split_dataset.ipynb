```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_onkk.csv')
df.head()
```

```python
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'])
df.head()
```

```python
df_sorted = df.sort_values(by='time').reset_index(drop=True)
df_sorted.head(20)
```

```python
n = len(df_sorted)
train_end = int(n * 0.7)
dev_end = train_end + int(n * 0.15)
```

```python
train_df = df_sorted.iloc[:train_end]
dev_df = df_sorted.iloc[train_end:dev_end]
test_df = df_sorted.iloc[dev_end:]
```

```python
train_df.head()
```

```python
dev_df.head()
```

```python
test_df.head()
```

```python
train_df.to_csv("train.csv", index=False)
dev_df.to_csv("dev.csv", index=False)
test_df.to_csv("test.csv", index=False)

# Kiểm tra thời gian trong mỗi tập
print("Train:", train_df['time'].min(), "→", train_df['time'].max())
print("Dev:", dev_df['time'].min(), "→", dev_df['time'].max())
print("Test:", test_df['time'].min(), "→", test_df['time'].max())
```
