```python
def open_and_view(path, title):
    # 1. Open with rasterio
    with rasterio.open(path) as src:
        arr = src.read(1)  # read the first band
        meta = src.meta
        nodata = src.nodata

    # 2. Display metadata and nodata
    print("Metadata:", meta)
    print("Nodata value:", nodata)

    # 3. Mask nodata in array
    arr_masked = np.where(arr == nodata, np.nan, arr)

    # 4. Print basic statistics
    print(f"Min: {np.nanmin(arr_masked):.4f}")
    print(f"Max: {np.nanmax(arr_masked):.4f}")
    print(f"Mean: {np.nanmean(arr_masked):.4f}")

    # 5. Plot the data
    plt.figure(figsize=(8, 6))
    plt.imshow(arr_masked, origin='upper')
    plt.title(title)
    plt.xlabel("Column (X)")
    plt.ylabel("Row (Y)")
    plt.colorbar(label=f"{title} Value")
    plt.show()
```

```python
# Path to the GeoTIFF file
path = root + r"\SQRT_SEA_DEM_LAT.tif"
open_and_view(path, "SQRT_SEA_DEM_LAT")
```
